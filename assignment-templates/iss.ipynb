## Libraries

```python
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
from datetime import datetime
import time as t
```

## Code and Graph

```python
lat = []
long = []
stamp = []

for num in range (0,101):
    url = "http://api.open-notify.org/iss-now.json"
    r = requests.get(url)
    iss_loc = r.json()
    if iss_loc['message'] == 'success':
        latitude = float(iss_loc['iss_position']['latitude'])
        longitude = float(iss_loc['iss_position']['longitude'])
        time = datetime.fromtimestamp(iss_loc['timestamp'])
        lat.append(latitude)
        long.append(longitude)
        stamp.append(time)
    t.sleep(10)

plt.plot(lat,long,'b.-')
plt.title("ISS Location",fontdict = {'fontweight' : 'bold','fontsize' : 18})
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```
